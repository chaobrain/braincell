# Morpho & Branch checkpoints

`braincell.io` ships a self-contained checkpoint format for snapshotting `Branch` and `Morpho` instances to disk and restoring them later. Two complementary persistence mechanisms are supported:

1. **`.bcm` checkpoints** — a stable, version-tagged on-disk format produced by `save_branch` / `save_morpho` (and the convenience methods `Branch.save_checkpoint` / `Morpho.save_checkpoint`). These are the recommended way to persist morphologies between sessions.
2. **`pickle`** — Python's standard serialization protocol works out of the box on `Branch`, `Morpho`, and `MorphoBranch` instances, including `copy.deepcopy`. Use it for in-process snapshots, multiprocessing, or wherever you would normally reach for `pickle`.

Both paths preserve information that the SWC/ASC interchange formats discard: branch *names*, `parent_x`/`child_x` attachments (including `parent_x=0.5` soma midpoints), the canonical zero-length jump segments inserted at radius discontinuities, and the auto-naming counters consulted by `Morpho.attach`.

This notebook walks through both mechanisms, when to pick one over the other, and how to handle the common error paths.

## 1. Setup

```python
import copy
import pickle
import tempfile
from pathlib import Path

import brainunit as u
import numpy as np

from braincell import (
    ApicalDendrite,
    Axon,
    BasalDendrite,
    Branch,
    Dendrite,
    Morpho,
    Soma,
)
from braincell.io import CheckpointError

# Use a temporary directory for everything in this notebook so it leaves no
# state behind on disk.
_tmp = tempfile.TemporaryDirectory()
TMP = Path(_tmp.name)
TMP
```

## 2. Saving and loading a single `Branch`

`Branch.save_checkpoint` writes the geometry and type to disk; `Branch.load_checkpoint` rebuilds an equivalent instance. The `.bcm` extension is appended automatically if you don't supply one.

```python
branch = Branch.from_lengths(
    lengths=[10.0, 15.0, 20.0] * u.um,
    radii=[3.0, 2.5, 2.0, 1.5] * u.um,
    type="dendrite",
)
branch
```

```python
checkpoint_path = branch.save_checkpoint(TMP / "my_branch")  # .bcm appended
checkpoint_path
```

```python
loaded = Branch.load_checkpoint(checkpoint_path)
print("equal to original?", loaded == branch)
print("loaded class:    ", type(loaded).__name__)
loaded
```

Note that the loaded class is `Dendrite` even though we constructed the original via `Branch.from_lengths(...)` directly. The checkpoint format stores the branch *type string* and reconstructs through `branch_class_for_type`, so loaded branches are always materialised as the appropriate typed subclass.

Equality (`==`) compares geometry plus type, so the upgraded subclass is still equal to the base-class original.

### Branches with 3-D point geometry

`Branch.from_points` produces branches with `points_proximal`/`points_distal` arrays. These are persisted as well.

```python
axon = Axon.from_points(
    points=np.array([
        [0.0, 0.0, 0.0],
        [10.0, 0.0, 0.0],
        [10.0, 20.0, 0.0],
        [10.0, 20.0, 30.0],
    ]) * u.um,
    radii=[2.0, 1.5, 1.0, 0.5] * u.um,
)
axon_path = axon.save_checkpoint(TMP / "axon.bcm")
loaded_axon = Branch.load_checkpoint(axon_path)
print("points preserved?", loaded_axon.points_proximal is not None)
print("equal?           ", loaded_axon == axon)
```

### Typed-subclass safety guard

When you call `load_checkpoint` on a typed subclass (e.g. `Soma.load_checkpoint`), the loader enforces that the saved type matches the subclass — analogous to the construction-time invariant. This catches accidental file mix-ups.

```python
# axon.bcm holds an Axon — Dendrite.load_checkpoint must refuse it.
try:
    Dendrite.load_checkpoint(axon_path)
except TypeError as exc:
    print("raised:", exc)
```

```python
# Use the base Branch loader (or Axon.load_checkpoint) when you don't want
# to assert a particular subclass.
Branch.load_checkpoint(axon_path)
```

## 3. Saving and loading a `Morpho`

Morphology checkpoints round-trip the entire tree: branch geometry, names, parent/child attachment points, and auto-naming counters. Build a tree by hand to demonstrate.

```python
soma = Soma.from_points(
    points=[[0.0, 0.0, 0.0], [20.0, 0.0, 0.0]] * u.um,
    radii=[10.0, 10.0] * u.um,
)
apical = ApicalDendrite.from_points(
    points=[[20.0, 0.0, 0.0], [20.0, 80.0, 0.0]] * u.um,
    radii=[2.0, 1.0] * u.um,
)
basal_a = BasalDendrite.from_points(
    points=[[0.0, 0.0, 0.0], [-30.0, -50.0, 10.0]] * u.um,
    radii=[3.0, 1.5] * u.um,
)
basal_b = BasalDendrite.from_points(
    points=[[0.0, 0.0, 0.0], [40.0, -40.0, -10.0], [60.0, -80.0, -20.0]] * u.um,
    radii=[2.5, 1.5, 0.8] * u.um,
)
axon_main = Axon.from_points(
    points=[[10.0, 0.0, 0.0], [10.0, 0.0, -100.0]] * u.um,
    radii=[1.0, 0.5] * u.um,
)

morpho = Morpho.from_root(soma, name="soma")
morpho.attach(parent="soma", child_branch=apical, child_name="apical", parent_x=1.0)
morpho.attach(parent="soma", child_branch=basal_a, child_name="basal_a", parent_x=0.0)
morpho.attach(parent="soma", child_branch=basal_b, child_name="basal_b", parent_x=0.0)
# parent_x=0.5 attaches to the midpoint of the soma — only legal on a soma
# parent and lossy under SWC/ASC, but checkpoints preserve it exactly.
morpho.attach(parent="soma", child_branch=axon_main, child_name="axon_main", parent_x=0.5)

print(morpho.topo())
morpho
```

```python
morpho_path = morpho.save_checkpoint(TMP / "my_morpho.bcm")
print("wrote", morpho_path)
print("size :", morpho_path.stat().st_size, "bytes")
```

```python
loaded = Morpho.load_checkpoint(morpho_path)
print("equal to original?", loaded == morpho)
print("topology preserved:")
print(loaded.topo())
print()
print("axon_main parent_x =", loaded.branch(name="axon_main").parent_x, "  # 0.5 soma midpoint")
print("basal_a parent_x   =", loaded.branch(name="basal_a").parent_x,   "  # 0.0 proximal")
```

### Auto-naming counters survive

`Morpho.attach` consults a per-type counter when generating default names like `dendrite_0`, `dendrite_1`, ... The checkpoint stores those counters so post-load attachments continue the same sequence.

```python
for _ in range(2):
    morpho.attach(
        parent="soma",
        child_branch=Dendrite.from_lengths(lengths=[10.0] * u.um, radii=[1.0, 0.5] * u.um),
    )
print("original auto-name counters:", morpho._type_name_counters)

snapshot_path = morpho.save_checkpoint(TMP / "with_counters.bcm")
reloaded = Morpho.load_checkpoint(snapshot_path)
print("reloaded auto-name counters:", reloaded._type_name_counters)

fresh = reloaded.attach(
    parent="soma",
    child_branch=Dendrite.from_lengths(lengths=[5.0] * u.um, radii=[1.0, 0.5] * u.um),
)
print("newly attached child auto-name:", fresh.name)
```

## 4. SWC / ASC → checkpoint snapshot workflow

A common pattern is to load a morphology from an interchange file once, do any cleanup or hand-editing, then snapshot the result as a checkpoint so subsequent sessions can skip the parsing/repair step entirely.

```python
swc_path = Path("morpho_files") / "branched_dend.swc"
if swc_path.exists():
    swc_morpho = Morpho.from_swc(swc_path)
    bcm_path = swc_morpho.save_checkpoint(TMP / swc_path.stem)
    print("snapshotted to:", bcm_path)
    print("branches      :", swc_morpho.n_branches)
    reloaded = Morpho.load_checkpoint(bcm_path)
    print("round-trip ok :", reloaded == swc_morpho)
else:
    print(f"{swc_path} not found — run this notebook from develop_doc/")
```

## 5. Pickle support

Every `Branch`, `Morpho`, and `MorphoBranch` instance is picklable using Python's stock `pickle` module. This means they also work with `copy.deepcopy`, `multiprocessing` queues, `joblib`, distributed task systems, and anything else that pickles its arguments under the hood.

```python
# Branch pickle round-trip.
pickled_branch = pickle.dumps(branch)
unpickled_branch = pickle.loads(pickled_branch)
print("branch equal?", unpickled_branch == branch)
print("branch class:", type(unpickled_branch).__name__)
print("pickle bytes:", len(pickled_branch))
```

```python
# Morpho pickle round-trip — back-references between MorphoBranch nodes and
# their owning Morpho are repaired automatically.
pickled = pickle.dumps(morpho)
unpickled = pickle.loads(pickled)
print("equal?         ", unpickled == morpho)
print("node count    :", len(unpickled))
print("back-ref ok   :", unpickled.soma._owner is unpickled)
print("attribute lookup still works:", unpickled.soma.apical.length)
```

```python
# copy.deepcopy uses the same protocol and works too.
deep_copy = copy.deepcopy(morpho)
print("is independent instance:", deep_copy is not morpho)
print("equal?                 :", deep_copy == morpho)
print("back-ref ok            :", deep_copy.soma._owner is deep_copy)
```

### Pickle vs `.bcm` — when to use which?

| Need | Use |
| --- | --- |
| Long-term on-disk archive | **`.bcm`** — version-tagged manifest, schema validation, friendly errors on corrupt input. |
| Sharing morphologies between collaborators or projects | **`.bcm`** — stable wire format independent of the in-memory layout. |
| Cross-version compatibility (load files saved by older braincell builds) | **`.bcm`** — the loader has explicit version checks and refuses unknown versions cloud `CheckpointVersionError`. |
| Multiprocessing / `joblib` / distributed workers | **`pickle`** — that's what those frameworks use under the hood. |
| Quick in-process clone | **`copy.deepcopy`** (which uses pickle). |
| Untrusted input | **`.bcm`** — the loader passes `allow_pickle=False`, so a malicious checkpoint cannot execute arbitrary Python. **Never** unpickle untrusted bytes. |

Pickle is faster for tiny morphologies because it skips the JSON manifest, but `.bcm` is the right choice for everything else. When in doubt, prefer `.bcm`.

## 6. Error handling

All checkpoint failures raise `CheckpointError` (a `ValueError` subclass), with `CheckpointVersionError` reserved for files written by a *newer* schema version than this build supports.

```python
# Loading a morpho file with the branch loader (or vice versa) raises a
# clear error pointing at the right loader.
try:
    Branch.load_checkpoint(morpho_path)
except CheckpointError as exc:
    print("raised:", exc)
```

```python
# A corrupt or non-checkpoint file is detected at the manifest layer.
garbage_path = TMP / "garbage.bcm"
garbage_path.write_bytes(b"this is not a zip archive")
try:
    Morpho.load_checkpoint(garbage_path)
except CheckpointError as exc:
    print("raised:", type(exc).__name__, "-", exc)
```
